# Convection Diffusion Problem With FEM

Let us have a look at how to solve convection diffusion problem using FEM, in particular let us focus on the following continuity equation,

$$ \partial_t \rho(t,\vec{x}) = \nabla \cdot \vec{J}(t,\vec{x})$$

we take as flux $\vec{J}(t,\vec{x})$ the combination of Fick's law and a convection term:

$$\vec{J}(t,\vec{x}) = -\epsilon\nabla \rho + \vec{p}(t,\vec{x})\rho(t,\vec{x})$$

plugining the above flux into the cotinuity eqaution one gets:

$$\partial_t \rho(t,\vec{x}) = -\epsilon\Delta \rho(t,\vec{x}) + \vec{p}(t,\vec{x})\cdot \nabla\rho(t,\vec{x}) + \rho(t,\vec{x})\nabla\cdot \vec{p}(t,\vec{x})$$

we now would like to add a source term to makes thing more interesting, and therefore we obtain:

$$\partial_t \rho(t,\vec{x}) = -\epsilon\Delta \rho(t,\vec{x}) + \vec{p}(t,\vec{x})\cdot \nabla\rho(t,\vec{x}) + \rho(t,\vec{x})\nabla\cdot \vec{p}(t,\vec{x})-f(t,\vec{x})$$

now we focus our intrest in the case when the "diffusive" process has reached a stationary regime, i.e. $\partial_t\rho \equiv 0$, imposing such a condition we obtain the __diffusion convection equation__:

 $$-\epsilon\Delta \rho(\vec{x}) + \vec{p}(\vec{x})\cdot \nabla\rho(\vec{x}) + \rho(\vec{x})\nabla\cdot \vec{p}(\vec{x})=f(\vec{x})$$
 
We'd like now to undesrtand how to solve the above mentioned PDE using numerical methods, to do this we focs on the 1D case, i.e.

$$-\epsilon\partial_x^2 \rho(x) + p(x)\partial_x \rho(x) + \rho(x)\partial_x p(x)=f(x)$$

let us introduce some concept that are gonna be useful to perform some analysis in this context.

__Definizione__ Si definisce $L^2(a,b)$ lo spazio vettoriale delle funzioni $u$ definite sull'intervallo $(a,b)$ tali che:

$$\int_a^b u^2(x) dx < \infty$$

lo spazio $L^2(a,b)$ è uno __spazio di Hilbert__ se munito della seguente norma e del seguente prodotto scalare:

$$||u||_{L^2(a,b)} = \Big( \int_a^b u^2(x) dx \Big)^{\frac{1}{2}}\qquad (u,w)_{L^2(a,b)} = \int_a^b u(x)w(x) dx $$

Possiamo dunque definire lo spazio $H^1(a,b)$ come lo spazio delle funzioni $u$ definite su $(a,b)$ tali che:

$$u\in L^2(a,b)\qquad \forall w \in C^\infty \textrm{ tale che } w(a)=w(b)=0 \Rightarrow \int_a^b w(x)\partial_xu(x) = \int_a^b u(x)\partial_xw(x)< \infty$$

lo spazio $H^1(a,b)$ è uno spazio di Hilbert rispetto alla seguente norma e al seguente prodotto scalare:

$$||u||_{H^1(a,b)}= ||u||_{L^2(a,b)} + ||\partial_x u||_{L^2(a,b)}\qquad (u,w)_{H^1(a,b)} = \int_a^b u(x)w(x) dx +\int_a^b \partial_x u(x) \partial_x w(x) dx $$

All'interno di $H^1(a,b)$ individuiamo il sotto spazio vettoriale $H^1_0(a,b) = \bigg\{ u\in H^1(a,b) \textrm{ tale che } u(a)=u(b)=0\bigg\}$.

Torniamo ora a considerare l'equazione di diffusione reazione convezzione ed osserviamo che se moltiplichiamo a destra e a sinistra per una funzione $w\in H^1_0(a,b)$ detta funzione test otteniamo:

$$-\epsilon \partial_x^2 \rho(x)w(x) + p(x)\partial_x \rho(x)w(x) + \rho(x)\partial_x p(x)w(x)$$

a questo punto integriamo per parti (e sostituiamo la $\rho$ con la $u$ al fine di mantenere la notazione della derivata debole), otteniamo:

$$\int_a^b \epsilon\partial_x u(x)\partial_x w(x) + q(x)u(x)w(x)-u(x)\partial_x w(x) dx = \int_a^b f(x)w(x) dx$$

a questo punti introduciamo la forma bilineare:

$$a(u,w) = \int_a^b \epsilon \partial_x u(x)\partial_x w(x) + q(x)u(x)w(x)-u(x)\partial_x w(x) dx $$

e la forma lineare $F(w)\int_a^b f(x)w(x) dx$ per poter scrivere il seguente problema variazionle:

$$ \textrm{Trovare u}\in H^1_0(a,b) \textrm{ tale che }a(u,w) = F(w) \qquad \forall w \in H^1_0(a,b)$$

per mostrare la buona posizione di questo problema varizionale si utilizza il seguente risultato:

>__Teorema (Lax-Milgram)__ Consideriamo il problema variazionale precedente, allora se valgono le seguenti ipotesi:
>1. $a(\cdot,\cdot)$ è __continua__, i.e. $\exists C_A$ tale che $\forall u,w \in H^1_0(a,b)$ si verifica $a(u,w)\leq ||u||_{H^1_0(a,b)}||w||_{H^1_0(a,b)}$;
>2. $a(\cdot,\cdot)$ è __fortemente coerciva__, i.e. $\exists \gamma_A$ tale che $\forall u \in H^1_0(a,b)$ si verifica $a(u,u)\geq ||u||^2_{H^1_0(a,b)}$;
>1. $F(\cdot)$ è __continua__, i.e. $\exists C_F$ tale che $\forall u \in H^1_0(a,b)$ si verifica $F(w)\leq ||w||_{H^1_0(a,b)}$;



E' venuto il momento di introdurre il metodo degli elementi finiti nel contesto dei metodi di Galerkin (continui), per fare questo per prima cosa introudicamo un sottospazio discreto $V^h\subset H^1_0(a,b)$; essendo $V_h$ uno spazio discreto possiamo individuare una base finita $\{\varphi_1,\dots, \varphi_N\}$. 
Torniamo ora al problema vairazionale astratto e osserviamo che perchè $a(u,w)=F(w)\;\forall w\in H^1_0(a,b)$ basta che valga sugli elementi della base:

$$a(u,\varphi_i)=F(\varphi_i)\;\forall 1 \leq i \leq N$$

scrivendo la soluzione $u$ nei termini della base discreta,i.e. $u=\sum_{k=1} U_k\varphi_k$ otteniamo:

$$a\bigg(\sum_{k=1}^{N} U_k\varphi_k,\varphi_i\bigg) = F(\varphi_i) \qquad \forall 1 \leq i \leq N$$

$$\sum_{k=1}^{N} U_k a(\varphi_k,\varphi_i) = F(\varphi_i) \qquad \forall 1 \leq i \leq N$$

questo equivale alla risoluzione del sistema lineare:

$$A\vec{U} = \vec{F} \qquad A_{k,i} = a(\varphi_k,\varphi_i),\;\vec{F}_{i}=F(\varphi_i)$$

per mostrare la convergenza del metodo numerico utilizziamo il seguente risultato:

>__Lemma (Cea)__ Consideriamo un problema variazionale astratto che verifica le ipotesi del teorema di Lax-Milgram e chiamiamo $u_h$ la soluzione del problema variazionale discreto:
>
>$$ \textrm{Trovare }u_h\in V^h \textrm{ tale che }a(u_h,w_h) = F(w_h) \qquad \forall w_h \in V^h$$
>
>allora vale la seguente stima per dell'errore:
>
>$$||u-u_h||_{H^1(a,b)} \leq \frac{C_A}{\gamma_A} \underset{w_h\in V^h}{\inf}||u-w_h||_V$$



Ora dobbiamo introdurre nel concreto lo spazio discreto $V^h$, negli elementi finiti si prende come spazio discreto lo spazio dei polinomi a tratti di grado $p\in N$ costruito sulla mesh $\mathcal{T}_h= \{K_j \textrm{ tale che } 1 \leq j \leq n+1\}$, dove $K_j = [a+(j-1)h,a+jh]$ prende il nome di __elemento__:
$$V^h = S^p(\mathcal{T}_h) = \bigg\{ w \in C^0(a,b) \textrm{ tale che } w_{|K_j} \in \mathbb{P}_p(K_j)\; \forall 1\leq j\leq n\bigg\}\qquad V^h_0 = S^p(\mathcal{T}_h)\cap H_0^1(a,b)$$
le dimensione degli spazi sopra menzionati sono date della seguenti formule:
$$dim\Big(S^p(\mathcal{T}_h)\Big) = (p+1)(n+1)-n\qquad dim\Big(S^p(\mathcal{T}_h)\cap H_0^1(a,b)\Big)=(p+1)(n+1)-n-2$$
Per prima cosa introdurremo gli __elementi finiti lineari__, i.e. $p=1$ e osserviamo che lo spazio $V^h_0$ in questo caso ha dimensione $2*(n+1)-n-2=n$, in particolare possiamo individuare la base di $V^h_0$ formata dalle funzioni a tenda:

$$ \varphi_j(x) = \begin{cases}
\frac{x-x_{j-1}}{x_j-x_{j-1}} \; in \; K_j\\
\frac{x_{j+1}-x}{x_{j+1}-x_{j}} \; in \; K_{j+1}\\
0 \; in \; (a,b)\backslash(K_j\cup K_{j+1})
\end{cases}$$

possiamo calcolare la derivate delle funzioni di base manualmente:

$$ \partial_x\varphi_j(x) = \begin{cases}
\frac{1}{x_{j}-x_{j-1}} \; in \; K_j\\
\frac{1}{x_{j+1}-x_{j}} \; in \; K_{j+1}\\
0 \; in \; (a,b)\backslash(K_j\cup K_{j+1})
\end{cases}$$

```python
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = 0.0; b=1; N = 10;
T = np.linspace(a,b,N); h = (b-a)/(N+1);
def tentbase(x,j,T):
    if T[j-1]<= x <= T[j]:
        return (x-T[j-1])/(T[j]-T[j-1]);
    elif T[j] <= x <= T[j+1]:
        return (T[j+1]-x)/(T[j+1]-T[j]);
    else:
        return 0;
def tentbasediff(x,j,T):
    if T[j-1]<= x <= T[j]:
        return 1/(T[j]-T[j-1]);
    elif T[j] <= x <= T[j+1]:
        return 1/(T[j+1]-T[j]);
    else:
        return 0;
X = np.linspace(a,b,100);
Y = [tentbase(x,2,T) for x in X];
plt.plot(X,Y)
X = np.linspace(a,b,100);
Y = [tentbase(x,5,T) for x in X];
plt.plot(X,Y)
plt.plot(T,np.zeros(len(T)),'r*')
```

Possiamo ora costruire la __matrice di stifness__ A, in particlare per semplicitò supporemo i termini di convezzione e di reazione siano costanti, ovvero q(x) = q e p(x) = p:

$$ A_{k,j} = \int_b^a \epsilon \varphi_k'\varphi_j' + q\varphi_k\varphi_j + p\varphi_k\varphi_j'$$

in particolare possiamo integrare a priori per ottenere dei termini prefissati per la diagonale e lo sovra e sotto diagonale:

$$A_{k,k} = \int_b^a \epsilon \varphi_k'\varphi_k' + q\varphi_k\varphi_k + p\varphi_k\varphi_k'=\frac{2\epsilon}{h}+\frac{2qh}{3}\qquad A_{k,k+1} = \int_b^a \epsilon \varphi_k'\varphi_{k+1}' + q\varphi_k\varphi_{k+1} + p\varphi_k\varphi_{k+1}'=-\frac{\epsilon}{h}+\frac{qh}{6}+\frac{p}{2}\qquad A_{k+1,k} = \int_b^a \epsilon\varphi_k'\varphi_{k+1}' + q\varphi_k\varphi_{k+1} + p\varphi_{k+1}\varphi_{k}'=-\frac{\epsilon}{h}+\frac{qh}{6}-\frac{p}{2}$$

gli altri elementi della matrice di $A$ saranno nulli inquanto i supporit delle funzioni di base non si intersecano e quindi gli integrali saranno nulli. Computiamo anche il vettore $F$ utilizzando come formula di quadratura quella del punto medio, i.e.

$$\vec{F}_j= \frac{hf_j+hf_{j+1}}{2}$$

__Esercizio__ Utilizzare il metodo agli elementi finiti per risolvere la seguente equazione di diffusione e reazione:

$$-u''+u = (1+\pi^2)\sin(\pi x)$$

```python
a = 0.0; b=1; N = 1000;
T = np.linspace(a,b,N); h = (b-a)/(N+1);
A = np.zeros((N,N));
eps = 1; q = 1; p = 0;
def f(x):
    return (1+np.pi**2)*np.sin(np.pi*x);
for k in range(0,N-1):
    A[k,k] = 2*eps/h+(2/3)*(q*h);
    A[k,k+1] = -eps/h+(1/6)*(q*h)-(p/2);
    A[k+1,k] = -eps/h+(1/6)*(q*h)+(p/2);
A[N-1,N-1] = 2*eps/h+(2/3)*(q*h);
F = np.zeros(N)
for k in range(0,N-1):
    F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
print(h)
U = np.linalg.solve(A,F);
U[0] = 0; U[-1]=0; #Boundary Condition
plt.plot(T,U,label="Soluzione Numerica")
X = np.linspace(a,b,100);
Y = [np.sin(np.pi*x) for x in X];
plt.plot(X,Y,label="Soluzione Esatta")
plt.legend()
```

__Osservazione__ Consideriamo ora un problema di diffusione trasporto generico:

$$\begin{cases}
-\epsilon u''(x)+ pu'(x) = 0\\
u(0) = 0\qquad u(1)=1
\end{cases}$$

questo particolare problema di Cauchy lo possiamo vedere come un equazione differenziale ordinaria del second'ordine e risolverla con il metodo degli esponenziali puri, ovvero per prima cosa calcoliamo le radici dell'equazione caratteristica:

$$-\epsilon\lambda^2+p\lambda = 0\qquad \lambda_{1,2}=\frac{p}{\epsilon}$$

allora possiamo trovare la solzuione nella combinazione lineare delle funzioni $u(x)=1$ e $u(x) = e^{\frac{p}{\epsilon}x}$ in particolare imponendo le soluzioni al bordo otteniamo:

$$ u(x) = \frac{e^{\frac{p}{\epsilon}x}-1}{e^{\frac{p}{\epsilon}}-1}$$

se il termine di diffusione è dominante i.e $\frac{p}{\epsilon} \approx 0$ possiamo sostituire i numeratore e denominatore con un sviluppo di Taylor attorno allo $0$ per osservare che, $ u(x) \approx x$ è la soluzione del problema di diffusione.

Nel caso in cui il termine di trasporto sia dominante le cose non sono così facili, inatti abbiamo che la soluzione è data da:

$$u(x)\approx e^{\frac{p}{\epsilon}(x-1)} $$

questa soluzione presenta uno slivellamento continuo attorno all' unità che risulta di difficle risoluzione numerica, pensando alla teoria delle equazioni differenziali ordinare otteniamo un problema di tipo stiff.

Risolviamo numericamente sia il caso in cui il termine di trasporto è dominante, sia quello in cui il termine di diffusione è dominante. Per imporre le condizioni al bordo osserviamo che data una funzione lineare f(x) detta lifting tale che $f(0) = u(0)$ e $f(1)=u(1)$ possiamo introdurre $v = u-f$ e risolvere il problema omogeneo:

$$\begin{cases}
-\epsilon v''(x)+ pv'(x) = -pf'(x)\\
v(0) = 0\qquad v(1)=1
\end{cases}$$

```python
a = 0.0; b=1; N = 40;
T = np.linspace(a,b,N); h = (b-a)/(N+1);
A = np.zeros((N,N));
eps = 10; q = 0; p = 0.1;
def f(x):
    return -p;
def lift(x):
    return x;
for k in range(0,N-1):
    A[k,k] = 2*eps/h+(2/3)*(q*h);
    A[k,k+1] = -eps/h+(1/6)*(q*h)-(p/2);
    A[k+1,k] = -eps/h+(1/6)*(q*h)+(p/2);
A[N-1,N-1] = 2*eps/h+(2/3)*(q*h);
F = np.zeros(N)
for k in range(0,N-1):
    F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
print(h)
U = np.linalg.solve(A,F);
U[0] = 0; U[-1]=0; #Boundary Condition
#Imponiamo condizioni al bordo non omogene mediante tecnica di lifting
for k in range(0,N):
    U[k] = U[k]+lift(k*h);
plt.plot(T,U,label="Soluzione Numerica")
plt.legend()
```

```python
a = 0.0; b=1; N = 40;
T = np.linspace(a,b,N); h = (b-a)/(N-1);
A = np.zeros((N,N));
eps = 0.1; q = 0; p = 1;
print("Numero di Peclet: ",p/eps)
def f(x):
    return -p;
def lift(x):
    return x;
for k in range(1,N-1):
    A[k,k] = +2*eps/h+(2/3)*(q*h);
    A[k,k+1] = -eps/h+(1/6)*(q*h)+(p/2);
    A[k+1,k] = -eps/h+(1/6)*(q*h)-(p/2);

F = np.zeros(N)
for k in range(1,N-1):
    F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
U[1:N-1] = np.linalg.solve(A[1:N-1,1:N-1],F[1:N-1]);
U[0] = 0; U[-1]=0; #Boundary Condition
#Imponiamo condizioni al bordo non omogene mediante tecnica di lifting
for k in range(0,N):
    U[k] = U[k]+lift(k*h);
plt.plot(T,U,label="Soluzione Numerica")
plt.legend()
```

In particolare dopo scelta molto piccola di $\epsilon$ abbiamo che il problema presenta delle oscillazioni spure, come possiamo osservare dalla seguente animazione.

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, 1))
ax.set_ylim((-1, 1))

line, = ax.plot([], [], lw=2)
plt.title("Diffusion Transport Equation")
def init():
    line.set_data([], [])
    return (line,)
def animate(j):
    a = 0.0; b=1; N = 40;
    T = np.linspace(a,b,N); h = (b-a)/(N-1);
    A = np.zeros((N,N));
    eps = 1/(10*j+1); q = 0; p = 1;
    def f(x):
        return -p;
    def lift(x):
        return x;
    for k in range(1,N-1):
        A[k,k] = +2*eps/h+(2/3)*(q*h);
        A[k,k+1] = -eps/h+(1/6)*(q*h)+(p/2);
        A[k+1,k] = -eps/h+(1/6)*(q*h)-(p/2);

    F = np.zeros(N)
    for k in range(1,N-1):
        F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
    U[1:N-1] = np.linalg.solve(A[1:N-1,1:N-1],F[1:N-1]);
    U[0] = 0; U[-1]=0; #Boundary Condition
    #Imponiamo condizioni al bordo non omogene mediante tecnica di lifting
    for k in range(0,N):
        U[k] = U[k]+lift(k*h);
    line.set_data(T, U)
    plt.title("Diffusion Transport Equation, Numero di peclet: "+str(p/eps)[0:5])
    return (line,)
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=True);
HTML(anim.to_html5_video())
```

Questo comportamento è congruo con quanto visto nel lemma di Cea, poichè la costante di coercività per la forma bilineare associata all'equazione di diffusione e trasporto è $\gamma_A=\mathcal{O}(\epsilon)$ dunque man man che $\epsilon$ diventa piccolo la stima dell' errore diventa sempre più lasca. Per questo problema ricorriamo a metodi di stabilizzazione per gli elementi finiti, dove invece che minimizzare la forma bilineare $a(\cdot,\cdot)$ minimizziamo la forma bilineare:

$$a(\cdot,\cdot)+c_h(\cdot,\cdot)$$

l'aggiunta della forma bilineare $c_h(\cdot,\cdot):H^1_0(a,b)\times H^1_0(a,b)\to \mathbb{R}$ punta a migliorare la coercività del problema. Inizieremo studiando i metodi di stabilizzazione non basati sul residuo dove $c_h(\cdot,\cdot)$ ha forma:

$$c_h(u,w) = \Big(Du,\tau Dw)$$

dove $D\cdot $ è un operatore differenziali preassegnato. La scelta di diversi operatori $D$ ci porta a diversi metodi, per esempio otteniamo i __metodi di diffusione artificiale__ prendendo:

$$D=\nabla \qquad \tau = h$$
$$c_h(u,w)=h(\nabla u,\nabla w)_{L^2(a,b)}=h\int_a^b \nabla u (x)\cdot\nabla v (x) dx=h\int_a^b \partial_x u(x) \partial_x v(x) dx$$

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, 1))
ax.set_ylim((-1, 1))

line, = ax.plot([], [], lw=2)
plt.title("Diffusion Transport Equation")
def init():
    line.set_data([], [])
    return (line,)
def animate(j):
    a = 0.0; b=1; N = 40;
    T = np.linspace(a,b,N); h = (b-a)/(N-1);
    A = np.zeros((N,N));
    eps = 1/(10*j+1)+h; q = 0; p = 1;
    def f(x):
        return -p;
    def lift(x):
        return x;
    for k in range(1,N-1):
        A[k,k] = +2*eps/h+(2/3)*(q*h);
        A[k,k+1] = -eps/h+(1/6)*(q*h)+(p/2);
        A[k+1,k] = -eps/h+(1/6)*(q*h)-(p/2);

    F = np.zeros(N)
    for k in range(1,N-1):
        F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
    U[1:N-1] = np.linalg.solve(A[1:N-1,1:N-1],F[1:N-1]);
    U[0] = 0; U[-1]=0; #Boundary Condition
    #Imponiamo condizioni al bordo non omogene mediante tecnica di lifting
    for k in range(0,N):
        U[k] = U[k]+lift(k*h);
    line.set_data(T, U)
    plt.title("Diffusion Transport Equation, Numero di peclet: "+str(p/(eps-h))[0:5])
    return (line,)
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=True);
HTML(anim.to_html5_video())
```

Un altra classe di metodi particolarmente importanti sono i metodi di __diffusione secondo corrente__:

$$D=\vec{p}\cdot\nabla \qquad \tau = h-\epsilon$$
$$c_h(u,w)=(h-\epsilon)(\vec{p}\cdot\nabla u,\vec{p}\cdot \nabla w)_{L^2(a,b)}=(h-\epsilon)\int_a^b p^2\partial_x u(x)\partial_x v(x) dx$$

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, 1))
ax.set_ylim((-1, 1))

line, = ax.plot([], [], lw=2)
plt.title("Diffusion Transport Equation")
def init():
    line.set_data([], [])
    return (line,)
def animate(j):
    a = 0.0; b=1; N = 40;
    T = np.linspace(a,b,N); h = (b-a)/(N-1);
    A = np.zeros((N,N));
    eps = 1/(10*j+1); q = 0; p = 1; K = eps+(h-eps)
    def f(x):
        return -p;
    def lift(x):
        return x;
    for k in range(1,N-1):
        A[k,k] = +2*K/h+(2/3)*(q*h);
        A[k,k+1] = -K/h+(1/6)*(q*h)+(p/2);
        A[k+1,k] = -K/h+(1/6)*(q*h)-(p/2);

    F = np.zeros(N)
    for k in range(1,N-1):
        F[k] = 0.5*h*(f(T[k])+f(T[k+1]));
    U[1:N-1] = np.linalg.solve(A[1:N-1,1:N-1],F[1:N-1]);
    U[0] = 0; U[-1]=0; #Boundary Condition
    #Imponiamo condizioni al bordo non omogene mediante tecnica di lifting
    for k in range(0,N):
        U[k] = U[k]+lift(k*h);
    line.set_data(T, U)
    plt.title("Diffusion Transport Equation, Numero di peclet: "+str(p/eps)[0:5])
    return (line,)
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=True);
HTML(anim.to_html5_video())
```

il problema di questi metodi di risoluzione agli elementi finiti è che sono __inconsistenti__ ovvero una soluzione del problema di diffusione-reazione-convezzione in formulazione variazionale non è soluzione anche del problema variazzionale stabilizzato. Infatti il problema variazionale stabilizzato è una perturbazione $\mathcal{O}(h)$ del problema originale.
